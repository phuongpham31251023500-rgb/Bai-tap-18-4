```python
import osmnx as ox
import networkx as nx
import pandas as pd
import folium

place_name = "District 10, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place_name, network_type="drive")
nodes,edges = ox.graph_to_gdfs(G)
center_lat = nodes.geometry.centroid.y.mean()
center_long = nodes.geometry.centroid.x.mean()


num_nodes = len(G.nodes)
num_edges = len(G.edges)
total_length = sum(data["length"] for u, v, k, data in G.edges(keys=True, data=True))
avg_length = total_length / num_edges
area_m2 = ox.geocoder.geocode_to_gdf(place_name).to_crs(epsg=3857).area.iloc[0]
area_km2 = area_m2 / 1e6
network_density = (total_length / 1000) / area_km2

result = pd.DataFrame({
    "Chỉ số": [
        "Số nút giao",
        "Số đoạn đường",
        "Tổng chiều dài đường (km)",
        "Chiều dài TB mỗi đoạn (m)",
        "Diện tích khu vực (km2)",
        "Mật độ mạng (km đường/km2)"
    ],
    "Giá trị": [
        num_nodes,
        num_edges,
        round(total_length/1000,2),
        round(avg_length,2),
        round(area_km2,2),
        round(network_density,2)
    ]
})

print(result)
m = folium.Map(location=[center_lat,center_long], zoom_start=15)
folium.GeoJson(edges).add_to(m)
m
```

```python
import osmnx as ox
import networkx as nx
import folium
from math import radians, sin, cos, sqrt, atan2

diem_dau = [10.761355552352372, 106.6683459252726]
diem_cuoi = [10.780211709469425, 106.69887295992045]

G = ox.graph_from_point(diem_dau, dist=6000, network_type='drive')

nut_dau = ox.distance.nearest_nodes(G, diem_dau[1], diem_dau[0])
nut_cuoi = ox.distance.nearest_nodes(G, diem_cuoi[1], diem_cuoi[0])

duong_dijkstra = nx.shortest_path(G, nut_dau, nut_cuoi, weight='length', method='dijkstra')

def khoang_cach_thang(lat1, lon1, lat2, lon2):
    R = 6371000
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2) ** 2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

def heuristic(u, v):
    lat1 = G.nodes[u]['y']
    lon1 = G.nodes[u]['x']
    lat2 = G.nodes[v]['y']
    lon2 = G.nodes[v]['x']
    return khoang_cach_thang(lat1, lon1, lat2, lon2)

duong_astar = nx.astar_path(G, nut_dau, nut_cuoi, heuristic=heuristic, weight='length')
def lay_toa_do(duong):
    toa_do = []
    for nut in duong:
        vi_do = G.nodes[nut]['y']
        kinh_do = G.nodes[nut]['x']
        toa_do.append([vi_do, kinh_do])
    return toa_do
m = folium.Map(location=diem_dau, zoom_start=13)
folium.Marker(diem_dau, popup='UEH', tooltip = 'click').add_to(m)
folium.Marker(diem_cuoi, popup='Nha tho Duc Ba',  tooltip = 'click').add_to(m)
folium.PolyLine(lay_toa_do(duong_dijkstra), color='blue', weight=5, popup='Dijkstra').add_to(m)
folium.PolyLine(lay_toa_do(duong_astar), color='red', weight=3, popup='A*').add_to(m)
m
```

```python
import folium
import osmnx as ox
import networkx as nx

khach_hang = [
    ['Khach 1', 10.761208381640975, 106.66835646578188],
    ['Khach 2', 10.783304116512499, 106.69462528061219],
    ['Khach 3', 10.773376523822351, 106.67750683643312],
    ['Khach 4', 10.790579315471804, 106.69939015362725]
]

xe = [
    ['Xe A', 10.762757567108883, 106.67021457267154],
    ['Xe B', 10.772706477635213, 106.69803150174114],
    ['Xe C', 10.795312129720845, 106.72206380944814],
    ['Xe D', 10.777488766355205, 106.69521071211855]
]

trung_tam = [10.765, 106.675]
G = ox.graph_from_point(trung_tam,dist=8000,network_type='drive')

def tim_nut(diem):
    return ox.distance.nearest_nodes(G, diem[2], diem[1])
def lay_toa_do(duong):
    toa_do = []
    for nut in duong:
        vi_do = G.nodes[nut]['y']
        kinh_do = G.nodes[nut]['x']
        toa_do.append([vi_do, kinh_do])
    return toa_do

ket_qua = []
xe_chua_ghep = xe.copy()

for khach in khach_hang:
    nut_khach = tim_nut(khach)
    xe_gan_nhat = None
    duong_ngan_nhat = None
    do_dai_nho_nhat = 999999999

    for tai_xe in xe_chua_ghep:
        nut_xe = tim_nut(tai_xe)
        try:
            do_dai = nx.shortest_path_length(
                G,
                nut_xe,
                nut_khach,
                weight='length'
            )
            duong = nx.shortest_path(
                G,
                nut_xe,
                nut_khach,
                weight='length',
                method='dijkstra'
            )
            if do_dai < do_dai_nho_nhat:
                do_dai_nho_nhat = do_dai
                xe_gan_nhat = tai_xe
                duong_ngan_nhat = duong

        except:
            pass
    if xe_gan_nhat != None:
        ket_qua.append([
            khach[0],
            xe_gan_nhat[0],
            round(do_dai_nho_nhat / 1000, 2),
            duong_ngan_nhat
        ])
        xe_chua_ghep.remove(xe_gan_nhat)

for dong in ket_qua:
    print(dong[0], 'duoc ghep voi', dong[1], '- quang duong', dong[2], 'km')

m = folium.Map(location=trung_tam, zoom_start=12)

for khach in khach_hang:
    folium.Marker(
        [khach[1], khach[2]],
        popup=khach[0],
        tooltip='Khach hang',
        icon=folium.Icon(color='red')
    ).add_to(m)
for tai_xe in xe:
    folium.Marker(
        [tai_xe[1], tai_xe[2]],
        popup=tai_xe[0],
        tooltip='Tai xe',
        icon=folium.Icon(color='green')
    ).add_to(m)
for dong in ket_qua:
    ten_khach = dong[0]
    ten_xe = dong[1]
    do_dai = dong[2]
    duong = dong[3]

    folium.PolyLine(
        lay_toa_do(duong),
        color='blue',
        weight=4,
        popup=ten_xe + ' den ' + ten_khach + ': ' + str(do_dai) + ' km'
    ).add_to(m)
m
```

```python
warehouses = {
   "Kho Q1": (10.7767, 106.7031),
   "Kho Tân Bình": (10.8016, 106.6520),
   "Kho Q7": (10.7290, 106.7216)
}

customers = {
   "Khách 1": (10.7900, 106.7100),
   "Khách 2": (10.7800, 106.7200),
   "Khách 3": (10.7700, 106.6800),
   "Khách 4": (10.7850, 106.7300),
   "Khách 5": (10.7650, 106.6900),
   "Khách 6": (10.8100, 106.6600),
   "Khách 7": (10.8150, 106.6400),
   "Khách 8": (10.7350, 106.7150),
   "Khách 9": (10.7200, 106.7300),
   "Khách 10": (10.7500, 106.6200)
}

center = (10.7767, 106.7031)
G = ox.graph_from_point(
   center,
   dist=12000,
   network_type="drive"
)


def nearest_node(point):
   lat, lon = point
   return ox.distance.nearest_nodes(G, lon, lat)

def road_distance(point_a, point_b):
   node_a = nearest_node(point_a)
   node_b = nearest_node(point_b)

   distance_m = nx.shortest_path_length(
       G,
       node_a,
       node_b,
       weight="length"
   )
   return distance_m / 1000
assigned = {}

for customer_name, customer_point in customers.items():
   nearest_warehouse = min(
       warehouses,
       key=lambda wh: road_distance(customer_point, warehouses[wh])
   )
   if nearest_warehouse not in assigned:
       assigned[nearest_warehouse] = {}
   assigned[nearest_warehouse][customer_name] = customer_point

def find_nearest_customer(current_point, remaining):
   nearest_name = None
   nearest_point = None
   nearest_distance = 999999

   for name, point in remaining.items():
       d = road_distance(current_point, point)

       if d < nearest_distance:
           nearest_distance = d
           nearest_name = name
           nearest_point = point

   return nearest_name, nearest_point, nearest_distance

def nearest_neighbor_route(start_point, customer_group):
   remaining = customer_group.copy()
   route = [("Kho", start_point)]
   current_point = start_point
   total_distance = 0

   while len(remaining) > 0:
       name, point, d = find_nearest_customer(current_point, remaining)

       route.append((name, point))
       total_distance += d
       current_point = point

       del remaining[name]

   total_distance += road_distance(current_point, start_point)
   route.append(("Kho", start_point))

   return route, total_distance

routes = {}
total_distance = 0

for warehouse_name, warehouse_point in warehouses.items():
   customer_group = assigned.get(warehouse_name, {})

   if len(customer_group) == 0:
       continue

   route, distance = nearest_neighbor_route(
       warehouse_point,
       customer_group
   )

   routes[warehouse_name] = route
   total_distance += distance

m = folium.Map(location=center, zoom_start=12)

colors = {
   "Kho Q1": "red",
   "Kho Tân Bình": "blue",
   "Kho Q7": "green"
}

for warehouse_name, warehouse_point in warehouses.items():
   folium.Marker(
       location=warehouse_point,
       popup=warehouse_name,
       icon=folium.Icon(color=colors[warehouse_name])
   ).add_to(m)

for warehouse_name, customer_group in assigned.items():
   for customer_name, customer_point in customer_group.items():
       folium.Marker(
           location=customer_point,
           popup=f"{customer_name} - {warehouse_name}"
       ).add_to(m)

for warehouse_name, route in routes.items():
   color = colors[warehouse_name]

   for i in range(len(route) - 1):
       point_a = route[i][1]
       point_b = route[i + 1][1]

       node_a = nearest_node(point_a)
       node_b = nearest_node(point_b)

       path = nx.shortest_path(
           G,
           node_a,
           node_b,
           weight="length"
       )

       path_coords = [
           (G.nodes[node]["y"], G.nodes[node]["x"])
           for node in path
       ]

       folium.PolyLine(
           path_coords,
           color=color
       ).add_to(m)
m
```

```python
import osmnx as ox
import networkx as nx
import folium
from IPython.display import display, clear_output
import time

center = (10.7767, 106.7031)
G = ox.graph_from_point(center,dist=5000,network_type="drive")

start_point = (10.7767, 106.7031)
end_point = (10.8016, 106.6520)

def nearest_node(point):
    lat, lon = point
    return ox.distance.nearest_nodes(G, lon, lat)

start_node = nearest_node(start_point)
end_node = nearest_node(end_point)

#tìm tuyến đường ngắn nhất
route = nx.shortest_path(G,start_node,end_node,weight="length")
route_coords = [
    (G.nodes[node]["y"], G.nodes[node]["x"])
    for node in route]


for t in range(len(route_coords)):
    current_position = route_coords[t]
    m = folium.Map(location=current_position, zoom_start=14)
    # vẽ toàn bộ tuyến
    folium.PolyLine(route_coords,color="blue",weight=4,opacity=0.6,popup="Tuyến đường").add_to(m)
    # điểm bắt đầu
    folium.Marker(start_point,popup="Điểm xuất phát",icon=folium.Icon(color="green")).add_to(m)
    # điểm kết thúc
    folium.Marker(end_point,popup="Điểm kết thúc",icon=folium.Icon(color="red")).add_to(m)
    # vị trí xe hiện tại
    folium.Marker(current_position,popup=f"Xe đang ở bước thời gian {t}"icon=folium.Icon(color="blue", icon="car", prefix="fa")).add_to(m)
    clear_output(wait=True)
    print(f"Bước thời gian: {t}")
    display(m)
    time.sleep(0.1)
# Vẽ đường đi thực tế
folium.PolyLine(locations=route_coords, color='green', weight=5, opacity=0.8).add_to(m)
m
```

```python
import numpy as np
import pandas as pd
import folium
from sklearn.cluster import KMeans

np.random.seed(42)
center = [10.7767, 106.7031]
n = 120
lat = np.random.uniform(10.70, 10.86, n)
lon = np.random.uniform(106.60, 106.77, n)
df = pd.DataFrame({"lat": lat,"lon": lon})

k = 4

X = df[["lat", "lon"]]

model = KMeans(n_clusters=k, random_state=42, n_init=10)
df["cluster"] = model.fit_predict(X)

centers = model.cluster_centers_

m = folium.Map(location=center, zoom_start=11)

colors = ["red", "blue", "green", "purple"]

for i in range(len(df)):
    cluster_id = int(df.loc[i, "cluster"])
    color = colors[cluster_id]

    folium.CircleMarker(
        location=[df.loc[i, "lat"], df.loc[i, "lon"]],
        radius=4,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=f"Khách hàng - Cụm {cluster_id + 1}"
    ).add_to(m)

for i, c in enumerate(centers):
    folium.Marker(
        location=[c[0], c[1]],
        popup=f"Điểm trung chuyển đề xuất {i+1}",
        icon=folium.Icon(color="black", icon="home")
    ).add_to(m)

m
```
